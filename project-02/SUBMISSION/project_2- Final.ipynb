## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, check for missing values (and replace them if it makes sense to do so). Explain your rationale for the way you choose to "impute" the missing data.

```python
# rename header for weeks
import re
for i in df.columns[7:]:
    result = re.sub('[^0-9]','', i) + " week"
    df.rename(columns={i: result}, inplace=True)

df.head()
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
# creates peaked column
peaked = df[df.columns[7:]].min(axis=1)
df.insert(7,"peaked", peaked)

# creates weeks on billboard
weeks_on = df[df.columns[7:]].count(axis=1)
df.insert(7,"weeks_on", weeks_on)

# convert dates into datetime
date_entered = []
for entered in df["date.entered"]:
    date_conv = datetime.strptime(entered, '%Y-%m-%d')
    date_entered.append(date_conv)
df["date.entered"] = date_entered

date_peaked = []
for entered in df["date.peaked"]:
    date_conv = datetime.strptime(entered, '%Y-%m-%d')
    date_peaked.append(date_conv)
df["date.peaked"] = date_peaked

# adds numbers till date peaked
weeks_to_peak = df["date.peaked"]-df["date.entered"]
df.insert(7,"weeks_to_peak", weeks_to_peak)
df["weeks_to_peak"]=(df["weeks_to_peak"].astype('timedelta64[D]'))/7

# unpivots weeks
df = pd.melt(df, id_vars=list(df.columns[0:10]), value_vars=list(df.columns[10:]))

#drops NaN
df = df.dropna(axis=0, how= 'any')

# coverts to rank to rank_value (101-1)
rank_value=[]
for rank in df.value:
    new_rank = 101-rank
    rank_value.append(new_rank)
    
df.insert(8,"rank_value", rank_value)

# coverts to peaked to peaked_value (101-1)
peaked_value=[]
for peaked in df.peaked:
    new_peaked = 101-peaked
    peaked_value.append(new_peaked)
df.insert(10,"peaked_value", peaked_value)
```
